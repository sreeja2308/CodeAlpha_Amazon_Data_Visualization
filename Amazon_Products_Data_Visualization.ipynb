```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/amazon.csv')
print(df.head())
```

```python
top_categories = df['category'].value_counts().head(10)
top_categories.plot(kind='barh')
plt.title("Top Product Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

```python
plt.hist(df['rating'], bins=10)
plt.title("Ratings Distribution")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.show()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)
plt.hist(df['discount_percentage'], bins=20)
plt.title("Discount Percentage Distribution")
plt.xlabel("Discount %")
plt.ylabel("Count")
plt.show()
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
plt.scatter(df['actual_price'], df['discounted_price'])
plt.title("Actual Price vs Discounted Price")
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")
plt.show()
```

```python
print("""
Conclusion:
1. Electronics and accessories dominate Amazon product listings.
2. Most products have high ratings.
3. Significant discounts are common on Amazon.
4. Discounted prices strongly relate to actual prices.
""")
```
